```python
import json
import matplotlib.pyplot as plt
```

```python
f = open("./news_DB.json")
```

```python
data = json.load(f)
```

```python
print(data[0]['tag'])
```

```python
cnt_tag = {}
for news in data:
    for tag in news['tag']:
        cnt_tag[tag] = cnt_tag.get(tag, 0) + 1
```

```python
list_tag = []
for key in cnt_tag:
    list_tag.append((key, cnt_tag[key]))
```

```python
list_tag.sort(key = lambda x: x[1], reverse=True)
```

```python
print(list_tag[0][1])
```
