```python
import json
import matplotlib.pyplot as plt
```

```python
f = open("./news_DB.json")
```

```python
data = json.load(f)
```

```python
print(data[0]['tag'])
```

```python
cnt_tag = {}
total = 0
for news in data:
    for tag in news['tag']:
        cnt_tag[tag] = cnt_tag.get(tag, 0) + 1
        total += 1
```

```python
list_tag = []
for key in cnt_tag:
    list_tag.append((key.upper(), cnt_tag[key]))
```

```python
list_tag.sort(key = lambda x: x[1], reverse=True)
```

```python
print(list_tag[0][1])
```

```python
width = 0.35

fig, ax = plt.subplots()

keys = []
values = []

for k,v in list_tag:
    keys.append(k)
    values.append(v)

fig.set_size_inches(18.5, 15)

print(keys[0])

ax.barh(keys[:20], values[:20], align='center', height=0.5)
```

```python
plt.show()
```
